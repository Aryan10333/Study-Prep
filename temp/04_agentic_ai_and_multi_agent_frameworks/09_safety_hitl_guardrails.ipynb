# 09 Safety, HITL & Guardrails: Confidence Gating Pipeline

**Scenario**: Implement a confidence-gating approval loop. If confidence falls below $0.85$, require human validation.

## Step 1: Gated Pipeline Setup

```python
import random

class ConfidenceGatedAgent:
    def __init__(self, confidence_threshold=0.85):
        self.threshold = confidence_threshold
        
    def execute_transaction(self, amount, user_confirmed=False):
        # Simulate model confidence generation
        confidence_score = round(random.uniform(0.70, 0.98), 2)
        print(f"\n[Transaction Request] Amount: ${amount} | Model Confidence: {confidence_score}")
        
        if confidence_score >= self.threshold:
            print(f"[AUTONOMOUS] Confidence {confidence_score} >= {self.threshold}. Transaction processed.")
            return f"Completed autonomously with confidence {confidence_score}."
            
        # If below threshold, trigger human approval gate (HITL)
        print(f"[GATED WARNING] Confidence {confidence_score} < {self.threshold}. Halting for human approval.")
        if user_confirmed:
            print("[HITL] Human approved transaction manually. Executing...")
            return f"Completed via HITL manual override."
        else:
            print("[HITL] human rejected or ignored transaction. Aborted.")
            return "Aborted: Human approval pending."

agent = ConfidenceGatedAgent(confidence_threshold=0.85)
# 1. High confidence autonomous execution run
random.seed(42)
print(agent.execute_transaction(amount=120))

print("-" * 40)
# 2. Low confidence manual approval run
random.seed(12)
print(agent.execute_transaction(amount=2500, user_confirmed=True))
```

## Output Explanation & Complexity Analysis

- **Escalation Gating**: The conditional check dynamically halts execution when confidence is low ($0.74 < 0.85$), verifying transaction safety rules are maintained.
- **Time Complexity**: $O(1)$ constant time check.
- **Space Complexity**: $O(1)$ memory requirements.
- **Component Denotations**:
  - The gating overhead is independent of sequence or loop iteration sizes.
