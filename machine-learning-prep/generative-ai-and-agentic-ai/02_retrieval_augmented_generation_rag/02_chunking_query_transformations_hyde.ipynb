## Technique 1: Parent-Child Chunking Strategy

```python
# Parent-Child Document Chunking & Multi-Query Expansion
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.prompts import ChatPromptTemplate

# Parent Splitter (Large Context) vs Child Splitter (Small Retrieval Units)
parent_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
child_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)

sample_text = "FlashAttention loads Q, K, V tiles directly into fast GPU SRAM. Standard self-attention materializes N x N attention matrices in slow High-Bandwidth Memory (HBM). By computing softmax online without writing intermediate tensors to HBM, FlashAttention achieves 2x-4x wall-clock speedups."

parent_chunks = parent_splitter.split_text(sample_text)
child_chunks = child_splitter.split_text(sample_text)

print(f"Parent Chunks: {len(parent_chunks)} | Child Chunks: {len(child_chunks)}")
```
