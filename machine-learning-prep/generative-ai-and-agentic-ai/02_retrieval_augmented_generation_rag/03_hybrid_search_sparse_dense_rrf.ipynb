## Technique 1: Reciprocal Rank Fusion (RRF) Implementation

```python
# Reciprocal Rank Fusion (RRF) Rank Aggregation Algorithm
import numpy as np

def reciprocal_rank_fusion(sparse_ranks: dict, dense_ranks: dict, k: int = 60) -> list[tuple[str, float]]:
    rrf_scores = {}
    all_docs = set(sparse_ranks.keys()).union(set(dense_ranks.keys()))
    
    for doc in all_docs:
        score = 0.0
        if doc in sparse_ranks:
            score += 1.0 / (k + sparse_ranks[doc])
        if doc in dense_ranks:
            score += 1.0 / (k + dense_ranks[doc])
        rrf_scores[doc] = score
        
    sorted_docs = sorted(rrf_scores.items(), key=lambda x: x[1], reverse=True)
    return sorted_docs

# Sample Retrieval Ranks
sparse_results = {"Doc_A": 1, "Doc_B": 2, "Doc_C": 5}
dense_results = {"Doc_C": 1, "Doc_B": 3, "Doc_D": 4}

combined_rrf = reciprocal_rank_fusion(sparse_results, dense_results, k=60)
print("Hybrid Search RRF Combined Ranks:")
for doc, score in combined_rrf:
    print(f"  {doc}: RRF Score = {score:.5f}")
```
