## Step 1: Environment Setup

```python
import os
from dotenv import load_dotenv

# Load API keys from repository root .env
load_dotenv()

# Create hidden vector database directory for local index persistence
os.makedirs(".vectordb", exist_ok=True)

print("=== Repository Environment Setup ===")
print("OPENAI_API_KEY present:", bool(os.getenv("OPENAI_API_KEY")))
print("GEMINI_API_KEY present:", bool(os.getenv("GEMINI_API_KEY")))
print("Hidden DB Directory Path:", os.path.abspath(".vectordb"))
```

## Step 2: Automated Pytest RAG Regression Assertion Suite

```python
def test_rag_retrieval_integrity():
    retrieved_chunks = ["Microservice A communicates with Database B via gRPC over TLS 1.3."]
    generated_answer = "Microservice A uses gRPC over TLS 1.3."
    
    # Assertion 1: Candidate chunks non-empty
    assert len(retrieved_chunks) > 0, "RAG Regression Failure: 0 chunks retrieved!"
    
    # Assertion 2: Fact grounded in context
    assert "TLS 1.3" in generated_answer, "RAG Regression Failure: Fact missing in answer!"
    
    print("=== Pytest Production RAG Regression Assertions Passed (2/2) ===")

test_rag_retrieval_integrity()
```

## Step 3: Executing Complete GenAI Debugged RCA LLM Call

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

if os.getenv("OPENAI_API_KEY"):
    context_sample = "Microservice A communicates with Database B via gRPC over TLS 1.3."
    prompt = ChatPromptTemplate.from_template("Answer question strictly using context:\nContext:\n{context}\n\nQuestion: {question}\nAnswer:")
    llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.0)
    
    response = llm.invoke(prompt.format(context=context_sample, question="What security standard is used?"))
    print("\n=== Complete Debugged RAG GenAI Response ===")
    print(response.content)
```
