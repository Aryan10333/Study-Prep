## Technique 1: Cross-Encoder Precision Reranking & Filtering

```python
# Cross-Encoder Reranking Simulation & Score Filtering
import numpy as np

class CrossEncoderReranker:
    def __init__(self, relevance_threshold: float = 0.70):
        self.threshold = relevance_threshold

    def rerank(self, query: str, candidate_docs: list[str]) -> list[dict]:
        # Simulated Cross-Encoder fine-grained relevance evaluation
        results = []
        for doc in candidate_docs:
            # Score based on exact query term overlap and semantic match
            score = 0.85 if "FlashAttention" in doc or "PagedAttention" in doc else 0.35
            results.append({"doc": doc, "score": score, "keep": score >= self.threshold})
        results.sort(key=lambda x: x["score"], reverse=True)
        return results

docs = [
    "Doc 1: FlashAttention bypasses HBM bandwidth bounds.",
    "Doc 2: The weather in San Francisco is foggy today.",
    "Doc 3: PagedAttention optimizes KV cache page allocation."
]

reranker = CrossEncoderReranker(relevance_threshold=0.70)
reranked = reranker.rerank("Explain FlashAttention memory optimizations", docs)

print("Cross-Encoder Reranking Results:")
for item in reranked:
    status = "KEPT" if item["keep"] else "DISCARDED"
    print(f"  [{status}] Score: {item['score']:.2f} | Text: '{item['doc']}'")
```
