# 08 Prompt Construction & Context Assembly

**Real-World Scenario**: Hospital Clinical Audit & Patient Chart Review Assistant.

This notebook demonstrates **Automated Source Citation Tag Injection** (`[Source 1: Patient_Chart_A.pdf, Page 3]`) and Grounded System Prompt enforcement.

## Step 1: Environment Setup

```python
import os
from dotenv import load_dotenv

# Load API keys from repository root .env
load_dotenv()

# Create hidden vector database directory for local index persistence
os.makedirs(".vectordb", exist_ok=True)

print("=== Repository Environment Setup ===")
print("OPENAI_API_KEY present:", bool(os.getenv("OPENAI_API_KEY")))
print("GEMINI_API_KEY present:", bool(os.getenv("GEMINI_API_KEY")))
print("Hidden DB Directory Path:", os.path.abspath(".vectordb"))
```

## Step 2: Automated Source Citation Injection Code

```python
patient_records = [
    {"doc": "Patient_Chart_1001.pdf", "page": 3, "text": "Patient administered 50mg Amoxicillin twice daily for sinus infection."},
    {"doc": "Intake_Form_1001.pdf", "page": 1, "text": "Patient has no known penicillin allergies."}
]

context_text = ""
for idx, r in enumerate(patient_records, 1):
    context_text += f"[Source {idx}: {r['doc']}, Page {r['page']}]
{r['text']}

"

system_prompt = f"""You are a Hospital Clinical Audit AI Assistant.
Answer the clinical audit question strictly using the provided context passages.
Always include exact source citation tags e.g. [Source 1: Patient_Chart_1001.pdf, Page 3] for every factual statement.

Context:
{context_text}

Question: What medication was prescribed to the patient and what dose?
Answer:"""

print("=== Assembled Grounded System Prompt Context ===")
print(system_prompt)
```
