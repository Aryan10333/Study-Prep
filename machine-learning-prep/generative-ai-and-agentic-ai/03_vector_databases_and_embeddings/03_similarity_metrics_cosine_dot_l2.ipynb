## Technique 1: Similarity Functions & Normalization Algebra

```python
# Vector Similarity Metric Functions & Normalization Math
import numpy as np

def cosine_similarity(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))

def dot_product(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.dot(a, b))

def euclidean_distance_l2(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.linalg.norm(a - b))

v_a = np.array([2.0, 1.0, 0.5])
v_b = np.array([1.5, 2.5, 0.0])

print(f"Cosine Similarity:  {cosine_similarity(v_a, v_b):.4f}")
print(f"Dot Product:        {dot_product(v_a, v_b):.4f}")
print(f"Euclidean L2 Dist:  {euclidean_distance_l2(v_a, v_b):.4f}")
```
