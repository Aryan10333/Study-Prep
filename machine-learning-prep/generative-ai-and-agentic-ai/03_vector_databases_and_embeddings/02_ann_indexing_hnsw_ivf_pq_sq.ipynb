## Technique 1: Product Quantization (PQ) Codebook Compression

```python
# Product Quantization (PQ) Compression Codebook Simulation
import numpy as np

class ProductQuantizer:
    def __init__(self, num_subvectors: int = 4, codebook_size: int = 256):
        self.m = num_subvectors
        self.k = codebook_size

    def compress_vector(self, vector: np.ndarray) -> list[int]:
        sub_dim = len(vector) // self.m
        quantized_indices = []
        for i in range(self.m):
            sub_vec = vector[i*sub_dim : (i+1)*sub_dim]
            # Assign dummy centroid index (0 to 255 -> 1 byte per subvector)
            centroid_idx = int(np.abs(np.mean(sub_vec) * 100)) % self.k
            quantized_indices.append(centroid_idx)
        return quantized_indices

# Execution
vector_384d = np.random.randn(384)
pq = ProductQuantizer(num_subvectors=4, codebook_size=256)
pq_code = pq.compress_vector(vector_384d)

raw_bytes = vector_384d.nbytes # 384 * 4 = 1536 bytes
pq_bytes = len(pq_code) # 4 bytes

print(f"Uncompressed Vector Size: {raw_bytes} bytes")
print(f"Product Quantized Size:   {pq_bytes} bytes")
print(f"Compression Ratio:        {raw_bytes / pq_bytes:.1f}x reduction")
```
