## Step 2: Production Prompt Regression Assertion Suite (`pytest`)

```python
import json

def run_prompt_unit_tests():
    raw_response = '{"ticket_id": "INC-9021", "status": "RESOLVED", "resolution_time_hrs": 2.5}'
    
    # Assertion 1: Valid JSON
    data = json.loads(raw_response)
    
    # Assertion 2: Required Schema Keys Present
    for key in ["ticket_id", "status", "resolution_time_hrs"]:
        assert key in data, f"Assertion Failed: Key '{key}' missing!"
        
    # Assertion 3: Type Validation
    assert isinstance(data["resolution_time_hrs"], (int, float)), "Assertion Failed: Time must be numeric!"
    
    print("=== All Prompt Regression Assertions Passed (3/3) ===")

run_prompt_unit_tests()
```
