## Technique 2: LLMLingua Perplexity Token Compression Engine

```python
import numpy as np

class LLMLinguaCompressor:
    def __init__(self, perplexity_threshold: float = 2.0):
        self.threshold = perplexity_threshold

    def compress_prompt(self, tokens: list[str], token_probabilities: list[float]) -> dict:
        perplexities = [1.0 / p for p in token_probabilities]
        compressed_tokens = [t for t, ppl in zip(tokens, perplexities) if ppl >= self.threshold]
        
        return {
            "compressed_prompt": " ".join(compressed_tokens),
            "original_count": len(tokens),
            "compressed_count": len(compressed_tokens),
            "reduction_rate": f"{(1.0 - len(compressed_tokens)/len(tokens))*100:.1f}%"
        }

sample_prompt_tokens = [
    "System:", "You", "are", "a", "helpful", "AI", "assistant.", 
    "Summarize", "the", "key", "financial", "metrics", "from", "the", "document."
]
simulated_probs = [0.10, 0.90, 0.95, 0.92, 0.85, 0.80, 0.75, 0.12, 0.91, 0.15, 0.11, 0.14, 0.88, 0.93, 0.18]

compressor = LLMLinguaCompressor(perplexity_threshold=2.0)
res = compressor.compress_prompt(sample_prompt_tokens, simulated_probs)
print(f"Compressed Prompt: '{res['compressed_prompt']}' ({res['reduction_rate']} reduction)")
```

## Part 3: Visualizing Token Perplexity Retention Profile

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')

perplexities = [1.0 / p for p in simulated_probs]
colors = ['#51cf66' if ppl >= 2.0 else '#ff6b6b' for ppl in perplexities]

plt.figure(figsize=(11, 4.5))
plt.bar(sample_prompt_tokens, perplexities, color=colors, alpha=0.85)
plt.axhline(2.0, color='#ff922b', linestyle='--', linewidth=2, label='Retention Threshold (PPL = 2.0)')

plt.title('LLMLingua Token Compression: Perplexity Entropy Profile')
plt.ylabel('Perplexity Score PPL(x_i)')
plt.xticks(rotation=45, ha='right', fontsize=9.5)
plt.legend(); plt.grid(True, axis='y'); plt.tight_layout(); plt.show()
```
