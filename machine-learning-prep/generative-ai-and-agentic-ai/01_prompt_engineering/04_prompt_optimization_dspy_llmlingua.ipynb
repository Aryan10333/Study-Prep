# 04 Prompt Optimization & Token Compression

Simulating LLMLingua perplexity-based token selection for prompt compression.

## Part 1: Perplexity Entropy Calculation & Compression Filter

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

tokens = ['System:', 'You', 'are', 'a', 'helpful', 'AI', 'assistant.', 'Answer', 'the', 'question:', 'What', 'is', 'gravity?']
perplexities = np.array([1.1, 1.2, 1.0, 1.1, 1.3, 2.1, 1.4, 7.8, 1.2, 2.0, 9.4, 1.1, 8.9])
threshold = 2.0

compressed_prompt = [t for t, ppl in zip(tokens, perplexities) if ppl > threshold]
print('Original Prompt:', ' '.join(tokens))
print('Compressed Prompt:', ' '.join(compressed_prompt))
print(f'Compression Ratio: {len(compressed_prompt)} / {len(tokens)} tokens ({len(compressed_prompt)/len(tokens):.1%})')
```

## Part 2: Visualizing Perplexity Retention Threshold

```python
colors = ['#51cf66' if ppl > threshold else '#ff6b6b' for ppl in perplexities]
plt.figure(figsize=(10, 4))
plt.bar(tokens, perplexities, color=colors)
plt.axhline(threshold, color='#ff922b', linestyle='--', label='Compression Threshold (PPL=2.0)')
plt.title('LLMLingua Token Perplexity Compression Profile')
plt.ylabel('Perplexity PPL(x_i)'); plt.xticks(rotation=45); plt.legend(); plt.grid(True, axis='y'); plt.show()
```
