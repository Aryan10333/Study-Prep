## Step 2: Prompt Caching Cost & Latency Calculator

```python
def calculate_prompt_caching_savings(num_queries=1000, sys_tokens=10000, user_tokens=1000):
    price_uncached = 3.00 / 1_000_000 # $3/1M
    price_cached = 0.30 / 1_000_000   # $0.30/1M (90% discount)
    
    uncached_cost = num_queries * (sys_tokens + user_tokens) * price_uncached
    cached_cost = (sys_tokens + user_tokens) * price_uncached + (num_queries - 1) * (sys_tokens * price_cached + user_tokens * price_uncached)
    
    savings_pct = (1 - cached_cost / uncached_cost) * 100
    
    print(f"=== Prompt Caching Economic Analysis ({num_queries:,} Queries) ===")
    print(f"Uncached Total Cost: ${uncached_cost:.2f}")
    print(f"Cached Total Cost:   ${cached_cost:.2f}")
    print(f"Net Cost Savings:    {savings_pct:.1f}%")

calculate_prompt_caching_savings()
```
