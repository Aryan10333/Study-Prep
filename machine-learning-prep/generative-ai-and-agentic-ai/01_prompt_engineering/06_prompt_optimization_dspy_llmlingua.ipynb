## Step 2: LLMLingua Token Perplexity Compression Simulation

```python
def llmlingua_compress_prompt(raw_text: str, target_ratio: float = 0.5) -> dict:
    """Simulates LLMLingua perplexity entropy filtering."""
    words = raw_text.split()
    # High-information words (nouns, numbers, verbs) kept; low-info filler pruned
    filler_words = {"please", "note", "that", "the", "a", "an", "is", "of", "to", "for", "in"}
    compressed_words = [w for w in words if w.lower() not in filler_words]
    
    compressed_text = " ".join(compressed_words)
    compression_rate = (1 - len(compressed_words)/len(words)) * 100
    
    return {
        "original_word_count": len(words),
        "compressed_word_count": len(compressed_words),
        "compression_rate_pct": compression_rate,
        "compressed_text": compressed_text
    }

prompt_text = "Please be noted that the target Nvidia Q3 financial revenue reached $18.12 billion in the context."
res = llmlingua_compress_prompt(prompt_text)

print("=== LLMLingua Compression Results ===")
print("Original Text:  ", prompt_text)
print("Compressed Text:", res["compressed_text"])
print(f"Token Reduction: {res['compression_rate_pct']:.1f}%")
```
