```python
import torch
import torch.nn.functional as F

def dpo_loss(policy_chosen_logps: torch.Tensor, policy_rejected_logps: torch.Tensor,
             ref_chosen_logps: torch.Tensor, ref_rejected_logps: torch.Tensor, beta: float = 0.1) -> torch.Tensor:
    pi_logratios = policy_chosen_logps - policy_rejected_logps
    ref_logratios = ref_chosen_logps - ref_rejected_logps
    logits = pi_logratios - ref_logratios
    return -F.logsigmoid(beta * logits).mean()

# Dummy Log-Probabilities
p_chosen = torch.tensor([-1.2]); p_rejected = torch.tensor([-2.8])
r_chosen = torch.tensor([-1.5]); r_rejected = torch.tensor([-2.2])

loss = dpo_loss(p_chosen, p_rejected, r_chosen, r_rejected, beta=0.1)
print(f"Direct Preference Optimization (DPO) Loss: {loss.item():.4f}")
```
