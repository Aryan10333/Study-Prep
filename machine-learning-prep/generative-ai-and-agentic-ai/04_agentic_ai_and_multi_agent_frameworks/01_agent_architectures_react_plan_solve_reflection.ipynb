## Technique 1: ReAct Agent Loop Execution Engine

```python
# ReAct Agent Loop Implementation from Scratch
class ReActAgent:
    def __init__(self, tools: dict):
        self.tools = tools
        self.trajectory = []

    def execute_tool(self, tool_name: str, query: str) -> str:
        if tool_name in self.tools:
            return self.tools[tool_name](query)
        return f"Error: Tool '{tool_name}' not found."

    def run_step(self, thought: str, action_tool: str, action_input: str) -> dict:
        self.trajectory.append(f"Thought: {thought}")
        self.trajectory.append(f"Action: {action_tool}({action_input})")
        
        observation = self.execute_tool(action_tool, action_input)
        self.trajectory.append(f"Observation: {observation}")
        
        return {"thought": thought, "action": action_tool, "observation": observation}

# Mock Tool Definition
def mock_search_tool(q: str) -> str:
    return "FlashAttention achieves 2x-4x speedups by tiling Q, K, V matrices into GPU SRAM."

agent = ReActAgent(tools={"search": mock_search_tool})
step1 = agent.run_step(
    thought="I need to search for FlashAttention GPU SRAM memory optimizations.",
    action_tool="search",
    action_input="FlashAttention SRAM"
)

print("Agent Trajectory:")
for line in agent.trajectory:
    print(f"  {line}")
```
