## Technique 1: Pydantic Schema Tool Binding & Execution

```python
# LangChain @tool Decorator & Pydantic Function Schema Binding
from langchain_core.tools import tool
from pydantic import BaseModel, Field

class StockQuerySchema(BaseModel):
    ticker: str = Field(description="Stock ticker symbol, e.g. NVDA, AAPL")
    days: int = Field(default=30, description="Historical range in days")

@tool(args_schema=StockQuerySchema)
def fetch_stock_price(ticker: str, days: int = 30) -> str:
    """Fetches historical stock prices for a given ticker."""
    return f"Ticker {ticker}: Last {days} days average price = $125.40."

print("LangChain Tool Name:", fetch_stock_price.name)
print("LangChain Tool Description:", fetch_stock_price.description)
print("Tool Execution Test:", fetch_stock_price.invoke({"ticker": "NVDA", "days": 7}))
```
