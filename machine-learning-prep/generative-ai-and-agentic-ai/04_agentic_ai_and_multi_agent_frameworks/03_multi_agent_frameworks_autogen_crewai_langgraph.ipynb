## Technique 1: Multi-Agent StateGraph State Machine Routing

```python
# Multi-Agent StateGraph State Machine Simulation (LangGraph Pattern)
class MultiAgentStateGraph:
    def __init__(self):
        self.state = {"messages": [], "next_agent": "Supervisor"}

    def supervisor_node(self, user_goal: str) -> str:
        self.state["messages"].append(f"[User Goal]: {user_goal}")
        if "search" in user_goal.lower() or "research" in user_goal.lower():
            self.state["next_agent"] = "Researcher"
        else:
            self.state["next_agent"] = "Coder"
        return self.state["next_agent"]

    def researcher_node(self) -> str:
        self.state["messages"].append("[Researcher Agent]: Found document on FlashAttention memory bounds.")
        self.state["next_agent"] = "FINISH"
        return "FINISH"

graph = MultiAgentStateGraph()
next_step = graph.supervisor_node("Please research FlashAttention performance")
print(f"Supervisor Routed Goal to: '{next_step}'")

final_step = graph.researcher_node()
print(f"Workflow Finished: '{final_step}'")
print("Full Message State History:")
for msg in graph.state["messages"]:
    print(f"  {msg}")
```
