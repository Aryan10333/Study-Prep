## Technique 1: Sliding Window Conversation Buffer & Summary Memory Truncation

```python
# Agent Memory Hierarchy: Conversation Buffer + Summary Truncation
class AgentMemoryBuffer:
    def __init__(self, max_buffer_turns: int = 2):
        self.max_turns = max_buffer_turns
        self.recent_history = []
        self.summary_memory = ""

    def add_turn(self, role: str, content: str):
        self.recent_history.append((role, content))
        if len(self.recent_history) > self.max_turns:
            # Move oldest turn to summary memory
            oldest_role, oldest_content = self.recent_history.pop(0)
            self.summary_memory += f" {oldest_role}: {oldest_content}"

    def get_prompt_context(self) -> str:
        return f"Summary Memory: {self.summary_memory.strip()}\nRecent Buffer: {self.recent_history}"

mem = AgentMemoryBuffer(max_buffer_turns=2)
mem.add_turn("User", "Hi, I am researching vLLM memory optimizations.")
mem.add_turn("Assistant", "vLLM uses PagedAttention to eliminate VRAM fragmentation.")
mem.add_turn("User", "What about Grouped-Query Attention?")

print(f"Summary Memory: '{mem.summary_memory}'")
print(f"Active Buffer:  {mem.recent_history}")
```
