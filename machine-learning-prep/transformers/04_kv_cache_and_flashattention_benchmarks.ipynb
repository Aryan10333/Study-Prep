# Transformer Notebook 04: KV Cache Footprints & FlashAttention Bottlenecks

This notebook details system memory calculations for **KV Caching** (MHA vs. MQA vs. GQA) and models the **Prefill vs. Decode Roofline Bottleneck**.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

def compute_kv_cache_gb(layers=80, kv_heads=8, head_dim=128, seq_len=4000, batch_size=1, precision_bytes=2):
    total_bytes = 2 * precision_bytes * layers * kv_heads * head_dim * seq_len * batch_size
    return total_bytes / (1024 ** 3)

vram_70b = compute_kv_cache_gb()
print(f'70B Model GQA (8 KV Heads, 4000 tokens, FP16) KV Cache VRAM: {vram_70b:.4f} GB')
```
