# Transformer Notebook 02: Positional Encodings & RoPE Math

This notebook explores the mathematical implementations and visual properties of **Sinusoidal Positional Encodings**, **Rotary Positional Embeddings (RoPE)**, and **NTK-Aware Context Length Extension**.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

def compute_sinusoidal_pe(max_pos=100, d_model=128):
    pe = np.zeros((max_pos, d_model))
    for pos in range(max_pos):
        for i in range(0, d_model, 2):
            div_term = 10000.0 ** (i / d_model)
            pe[pos, i] = np.sin(pos / div_term)
            pe[pos, i + 1] = np.cos(pos / div_term)
    return pe

pe = compute_sinusoidal_pe()
plt.figure(figsize=(10, 4))
plt.imshow(pe, cmap='magma', aspect='auto')
plt.title('Sinusoidal Positional Encoding Heatmap (100 positions x 128 dims)')
plt.xlabel('Embedding Dimension'); plt.ylabel('Position Index')
plt.colorbar()
plt.show()
```

## RoPE 2D Vector Rotation Logic

RoPE rotates 2D slices of Query/Key vectors by an angle $\phi = m \theta$ proportional to the token position index $m$.

```python
def rotate_2d_vector(v, pos, theta=np.pi/6):
    angle = pos * theta
    R = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
    return R @ v

v = np.array([1.0, 0.0])
for pos in range(4):
    v_rot = rotate_2d_vector(v, pos)
    print(f'Position {pos} Rotated Vector: {np.round(v_rot, 4)}')
```
