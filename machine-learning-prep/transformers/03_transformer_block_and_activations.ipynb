# Transformer Notebook 03: Transformer Blocks, Activations & Pre-LN Stability

This notebook benchmarks activation functions (**ReLU**, **GELU**, **SwiGLU**) and simulates backpropagation gradient flow across deep Transformer layers under **Pre-LN** vs. **Post-LN** layouts.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
x = np.linspace(-3, 3, 300)
relu = np.maximum(0, x)
gelu = 0.5 * x * (1.0 + np.tanh(np.sqrt(2.0 / np.pi) * (x + 0.044715 * (x ** 3))))
swish = x / (1.0 + np.exp(-x))

plt.figure(figsize=(9, 4.5))
plt.plot(x, relu, label='ReLU', linestyle='--')
plt.plot(x, gelu, label='GELU (Smooth, non-zero negative gradients)', lw=2)
plt.plot(x, swish, label='Swish (SwiGLU Gate)', lw=2)
plt.title('Comparison of Activation Functions')
plt.xlabel('Input Activation x'); plt.ylabel('Output f(x)')
plt.grid(True); plt.legend()
plt.show()
```
