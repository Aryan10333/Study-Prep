# 03 Self-Attention Layer (PyTorch Module & Matrix Visualizer)

Implementing a custom PyTorch `CustomSelfAttention` layer with learnable linear projection weight matrices ($W_Q, W_K, W_V$).

## Part 1: PyTorch Self-Attention Module Definition
Constructing the linear projection layers and vectorized scaled attention logic.

```python
import torch
import torch.nn as nn
import math
import matplotlib.pyplot as plt

plt.style.use('dark_background')

class CustomSelfAttention(nn.Module):
    def __init__(self, d_model, d_k):
        super().__init__()
        self.d_k = d_k
        self.w_q = nn.Linear(d_model, d_k, bias=False)
        self.w_k = nn.Linear(d_model, d_k, bias=False)
        self.w_v = nn.Linear(d_model, d_k, bias=False)
        
    def forward(self, x):
        q = self.w_q(x)
        k = self.w_k(x)
        v = self.w_v(x)
        scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(self.d_k)
        attn_weights = torch.softmax(scores, dim=-1)
        return torch.matmul(attn_weights, v), attn_weights

x = torch.randn(1, 8, 128)
attn_layer = CustomSelfAttention(d_model=128, d_k=32)
out, weights = attn_layer(x)
print('PyTorch Self-Attention Output Shape:', out.shape)
```

## Part 2: Visualizing Computed PyTorch Attention Matrix
Plotting the attention probability heatmap across sequence tokens.

```python
plt.figure(figsize=(6, 5))
plt.imshow(weights.squeeze().detach().numpy(), cmap='viridis')
plt.title('PyTorch Self-Attention Weight Matrix (8 Tokens x 8 Tokens)')
plt.xlabel('Key Index'); plt.ylabel('Query Index')
plt.colorbar()
plt.show()
```
